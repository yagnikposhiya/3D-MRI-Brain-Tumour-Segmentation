## Loss graph plotting

````python
font = {'family':'serif', 'color':'#05445E', 'weight':'normal', 'size':14}
font_legend = {'family':'serif', 'weight':'normal', 'size':10}

# plt.figure(figsize=(6.8,4.8)): only for precision

plt.plot(dataframe_training['Epoch'], dataframe_training['Loss'], label='Training Loss', color='#FF8300')
plt.plot(dataframe_validation['Epoch'], dataframe_validation['Loss'], label='Validation Loss', color='#333652')
plt.xlim([0,36])
plt.ylim([0,0.35])

plt.xlabel('Epoch', fontdict=font)
plt.ylabel('Loss', fontdict=font)

L = plt.legend(loc='upper right', prop=font_legend)
for text in L.get_texts():
    text.set_color('#05445E')

plt.xticks(fontname='serif')
plt.yticks(fontname='serif')

plt.scatter(dataframe_training['Epoch'][1],dataframe_training['Loss'][1], color='#FF8300')
plt.scatter(dataframe_training['Epoch'][35],dataframe_training['Loss'][35], color='#FF8300')
plt.scatter(dataframe_validation['Epoch'][1],dataframe_validation['Loss'][1], color='#333652')
plt.scatter(dataframe_validation['Epoch'][35],dataframe_validation['Loss'][35], color='#333652')

plt.hlines(dataframe_training['Loss'][1],0,1, linestyles='dotted',  alpha=0.48, colors='#B68D40')
plt.hlines(dataframe_training['Loss'][35],0,35, linestyles='dotted', alpha=0.48, colors='#B68D40')
plt.hlines(dataframe_validation['Loss'][1],0,1, linestyles='dotted',  alpha=0.48, colors='#B68D40')
plt.hlines(dataframe_validation['Loss'][35],0,35, linestyles='dotted', alpha=0.48, colors='#B68D40')

plt.vlines(1,0, dataframe_training['Loss'][1], linestyles='dotted',  alpha=0.48, colors='#B68D40')
plt.vlines(35,0, dataframe_training['Loss'][35], linestyles='dotted', alpha=0.48, colors='#B68D40')
plt.vlines(1,0,dataframe_validation['Loss'][1], linestyles='dotted',  alpha=0.48, colors='#B68D40')
plt.vlines(35,0,dataframe_validation['Loss'][35], linestyles='dotted', alpha=0.48, colors='#B68D40')

plt.annotate('({start},{end})'.format(start=1,end=dataframe_training['Loss'][1]), xy=(1, dataframe_training['Loss'][1]), xytext=(2, dataframe_training['Loss'][1]-0.005), fontname='serif', color='#FF8300')
plt.annotate('({start},{end})'.format(start=35,end=dataframe_training['Loss'][35]), xy=(35, dataframe_training['Loss'][35]), xytext=(27.78, dataframe_training['Loss'][35]-0.0127), fontname='serif', color='#FF8300')
plt.annotate('({start},{end})'.format(start=1,end=dataframe_validation['Loss'][1]), xy=(1, dataframe_validation['Loss'][1]), xytext=(2, 0.082), fontname='serif', color='#333652')
plt.annotate('({start},{end})'.format(start=35,end=dataframe_validation['Loss'][35]), xy=(35, dataframe_validation['Loss'][35]), xytext=(27.78, 0.035), fontname='serif', color='#333652')
# plt.scatter(0,dataframe_training['Loss'][1])
plt.savefig('/home/yagnikposhiya/WorkStation/Lab/3MBTS/Training Validation epoch vs loss.png', format='png', dpi=2400)
````
